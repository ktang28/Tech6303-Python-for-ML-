```python
# read the data into a dataframe
import pandas as pd

file_path = 'C:/Users/ktang28/ML_Mini_project1_Chocolate/flavors_of_cacao.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#5.an output plot of the histogram of the values in the column named “Ratings;”.
import matplotlib.pyplot as plt
ratings = df.Rating.values

plt.figure(figsize=(20,5))
plt.hist(ratings, bins = 4, color = 'red')

plt.xlabel("Ratings")
plt.title("Chocolate Rating Since 2006")
```

```python
# remove non-digit characters from the ‘CocoaPercent’ column and convert the values to integers
df['CocoaPercent'] = df['CocoaPercent'].astype(str)
df['CocoaPercent'] = df['CocoaPercent'].str.replace(r'\D', '').astype(int)
#df['CocoaPercent'] = df['CocoaPercent'].str.replace(r'\D', '', regex=True).astype(int)
df
```

```python
# # 6. scatter plot between the cocoa percent values and rating values

cocoaPer = df.CocoaPercent.values
ratings = df.Rating.values

plt.figure(figsize=(20,5))
plt.scatter(cocoaPer,ratings, color = 'red')

plt.xlabel("Chocolate Percentage")
plt.title("Chocolate Percentage vs Rating Since 2006")
```
